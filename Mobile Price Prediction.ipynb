```python
X = df.drop('Price',axis=1)
y = df['Price']
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.2, random_state = 7)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
l_model  = LinearRegression()
l_model.fit(X_train,y_train)
y_pred = l_model.predict(X_test)
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print(f"R2_Score is: {r2_score(y_test,y_pred)}")
print(f"Mean Absolute error is: {mean_absolute_error(y_test,y_pred)}")
print(f"Mean Squared error is: {mean_squared_error(y_test,y_pred)}")
print(f"Root Mean Squared Error is: {np.sqrt(mean_squared_error(y_test,y_pred))}")
```

```python
from xgboost import XGBRegressor
xgb_model = XGBRegressor()
xgb_model.fit(X_train,y_train)
```

```python
y_pred_xgb = xgb_model.predict(X_test)
print(f"R2_Score is: {r2_score(y_test,y_pred_xgb)}")
print(f"Mean Absolute error is: {mean_absolute_error(y_test,y_pred_xgb)}")
print(f"Mean Squared error is: {mean_squared_error(y_test,y_pred_xgb)}")
print(f"Root Mean Squared Error is: {np.sqrt(mean_squared_error(y_test,y_pred_xgb))}")
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train,y_train)
```

```python
y_pred_dt = dt_model.predict(X_test)
print(f"R2_Score is: {r2_score(y_test,y_pred_dt)}")
print(f"Mean Absolute error is: {mean_absolute_error(y_test,y_pred_dt)}")
print(f"Mean Squared error is: {mean_squared_error(y_test,y_pred_dt)}")
print(f"Root Mean Squared Error is: {np.sqrt(mean_squared_error(y_test,y_pred_dt))}")
```
